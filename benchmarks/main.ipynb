```python
# Suponha que voc tenha trs DataFrames
df1 = pd.DataFrame({"A": [1, 4, 3], "B": [2, 3, 5], "C": [3, 2, 1]})

df2 = pd.DataFrame({"A": [3, 1, 2], "B": [4, 2, 3], "C": [1, 5, 4]})

df3 = pd.DataFrame({"A": [2, 3, 1], "B": [1, 4, 2], "C": [5, 1, 3]})

# Combine os DataFrames pegando o maior valor de cada linha correspondente
result = pd.concat([df1, df2, df3]).groupby(level=0).max()

print(result)
```

```python
evaluation_t1_1p = evaluation_t1_1p_pA[["image_id", "time_diff_ms"]]
evaluation_t1_2p = (
    pd.concat(
        [
            evaluation_t1_2p_pA[["image_id", "time_diff_ms"]],
            evaluation_t1_2p_pB[["image_id", "time_diff_ms"]],
        ]
    )
    .groupby(level=0)
    .max()
)
evaluation_t1_3p = (
    pd.concat(
        [
            evaluation_t1_3p_pA[["image_id", "time_diff_ms"]],
            evaluation_t1_3p_pB[["image_id", "time_diff_ms"]],
            evaluation_t1_3p_pC[["image_id", "time_diff_ms"]],
        ]
    )
    .groupby(level=0)
    .max()
)
```
